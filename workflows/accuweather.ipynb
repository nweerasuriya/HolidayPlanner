```python
importlib.reload(sys.modules['src.methods.weather_api'])

lat = 52.9540
long = 1.1550
# get weather data
loop = asyncio.new_event_loop()
current_conditions, forecast_daily, forecast_hourly = loop.run_until_complete(weather_api.main(lat, long))
loop.close()
```

```python
current_dict = current_conditions.copy()
forecast_daily_list = forecast_daily.copy()
forecast_hourly_dict = forecast_hourly.copy()
```

```python
# For nested dictionary only get metric key
for key, value in current_dict.items():
    if isinstance(value, dict):
        if 'Metric' in value.keys():
            current_dict[key] = value['Metric']
        else:
            for k, v in value.items():
                if isinstance(v, dict):
                    if 'Metric' in v.keys():
                        current_dict[key][k] = v['Metric']

current_dict
```

```python
# convert current conditions to dataframe
current_conditions_df = pd.DataFrame.from_dict(current_dict, orient='index').T
current_conditions_df
```
